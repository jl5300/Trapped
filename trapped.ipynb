```python
def first_room():

    print("You wake up on a cold floor to the sound of your dry coughing.")

    time.sleep(2)

    print("""
You don't recognize the room you're in.
It's illuminated dimly by a single light on the ceiling.""")

    time.sleep(2)

    print("You stand up.\n")

    print("""
There are three doors lining the wall across from you.
The room has no windows.
    """)

    time.sleep(2)

    first_room_ask_which_door()
        
```

```python
def first_room_ask_which_door():

    print("Which door do you approach? Type 1, 2, or 3.")
    door_choice = input()

    while(door_choice != "1" and door_choice != "2" and door_choice != "3"):
        print("Invalid input. Try again.")
        door_choice = input()

    if(door_choice == "1"):
        print("You hear the sound of dogs barking and chains rattling on the other side of the door.")

        time.sleep(2)

        print("Do you still want to enter? Type yes to enter, or no to return.")
        choice = input()

        while(choice.lower() != "yes" and choice.lower() != "no"):
            print("Invalid input. Try again.")
            choice = input()

        if(choice.lower() == "yes"):
            enter_room("1")
            
        else:
            print("You return to the center of the room.")
            first_ask_which_door()

    if(door_choice == "2"):

        print("You grab the doorknob to turn it. It's hot.")
        print("Do you still want to enter? Type yes to enter, or no to return.")
        choice = input()

        while(choice.lower() != "yes" and choice.lower() != "no"):
            print("Invalid input. Try again.")
            choice = input()

        if(choice.lower() == "yes"):
            print("A fire blazes out from the door as soon as you open it. You burn to death.")
            game_over()

        else:
            print("You return to the center of the room.")
            first_ask_which_door()

    if(door_choice == "3"):    
        print("You approach the door and place your ear to it. You don't hear anything.")
        print("Do you still want to enter? Type yes to enter, or no to return.")
        choice = input()

        while(choice.lower() != "yes" and choice.lower() != "no"):
            print("Invalid input. Try again.")
            choice = input()
        
        if(choice.lower() == "yes"):
            print("You open the door. The room is pitch dark.")

            time.sleep(2)

            print("Nervous, you try to turn back.")
            
            time.sleep(1)

            print("As you turn, your ankle is grabbed and you fall to the ground.\nYou are pulled into the dark room.")

            time.sleep(1)
            
            print("You try to scream. It's too late.")

            game_over()

        else:
            print("You return to the center of the room.")
            first_ask_which_door()
```

```python
def first_ask_which_door(xy):

    print("There are four doors in the room. Which way do you go? Type north, south, east, or west.")
    door_choice = input()
    door_choice = door_choice.lower()

    while(door_choice != "north" and door_choice != "south" and door_choice != "east" and door_choice != "west"):
        print("Invalid input. Try again.")
        door_choice = input()

    while(door_choice == "south"):
        print("You try to open the door behind you. It's locked.")
        first_ask_which_door(xy)

    if(door_choice == "north"):
        xy[1] += 1

    if(door_choice == "west"):
        xy[0] += 1

    if(door_choice == "east"):
        xy[0] -= 1
        
    print("You move " + door_choice + ".")
    return xy
```

```python
def ask_which_door(xy):

  door_choice = input("There are four doors in the room, one on each wall. Which way do you go? Type north, south, east, or west.")
  door_choice = door_choice.lower()

  while(door_choice != "north" and door_choice != "south" and door_choice != "east" and door_choice != "west"):
      print("Invalid input. Try again.")
      door_choice = input()

  if(door_choice == "north"):
      xy[1] += 1

  if(door_choice == "south"):
      xy[1] -= 1

  if(door_choice == "east"):
      xy[0] += 1

  if(door_choice == "west"):
      xy[0] -= 1

  print("You move " + door_choice + ".")
  return xy
```
